## Plot multiple stdf

### 1. Query

```python
from fetch import QueryParam, Query

extra_pipeline: ["group:min", "get_pout_from_text"]
value_min: -100,
test_text = ".*_rssi$",
plot_type = "point",
x_axis = "pout",
y_axis = "rssi_min",


DB_NAME = "firecrest"   # all previously saved data can be pulled by filters (stdf_name and test_text)
EXTRA_PIPELINE = "gt:-100+rssi_min"

query_params = [
    QueryParam(
        stdf_name="fc_a1_qfn_wifi_shmoo_min_svc40_rssi_site8_run1_20220825185348.stdf.gz",  # stdf should be in the same folder with this notebook 
        text="a_evmXrx_11axmcs0_5855_x_x_x_x_x_rssi",  # _x_ is wildcard
        stdf_id="svc40",   # stdf_id is for display only, not saved in data base
    ),
    QueryParam(
        stdf_name="fc_a1_qfn_wifi_shmoo_min_svc_rssi_mcs4_20220826124302.stdf.gz",
        text="a_evmXrx_11axmcs4_5855_x_x_x_x_x_rssi",
        stdf_id="svc",
    ),
]

df = Query(DB_NAME).get_df(query_params, EXTRA_PIPELINE)
df.head()


# use stdf folder name as db_name

```

### 2. Plot
#### 2.1 Overlay all

```python
from lets_plot import *
from lets_plot.mapping import as_discrete
LetsPlot.setup_html()

def plot(x, y):
    ggplot(data=df) \
        + geom_point(mapping = aes(x='pout', y='rssi_min', color='key')) \
        + scale_color_discrete() \
        + ggtitle("Overlay all") \
        + theme(legend_position=[-1, -1])
```

#### 2.2 Loop count split

```python
ggplot(data=df[df["stdf_id"] == "svc40"]) \
    + geom_point(mapping = aes(x='pout', y='rssi_min', color='key')) \
    + scale_color_discrete() \
    + ggtitle("svc40_5855_mcs0", subtitle="20 parts") \
    + facet_wrap(facets="part_id", format="loop {}0") \
    + theme(legend_position=[-1, -1])
```

#### 2.3 svc_5855_mcs4

```python
ggplot(data=df[df["stdf_id"] == "svc"]) \
    + geom_point(mapping = aes(x='pout', y='rssi_min', color='key')) \
    + scale_color_discrete() \
    + ggtitle("svc_5855_mcs4") \
    + theme(legend_position=[-1, -1])
```
